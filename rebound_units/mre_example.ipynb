```python
import rebound
import numpy as np
import matplotlib.pyplot as plt
```

### Let's start by initializing a planet 1 AU from a star of 1 Solar mass, with velocity = 2 $\pi$ AU/yr.  We expect this body to have a semi major axis of one, and a circular orbit.

#### First get the orbital elements using the sim.units 

```python
sim = rebound.Simulation()
sim.units=("Yr","Au","Msun")
sim.add(m=1,x=0,y=0,vx=0,vy=0)
sim.add(m=0,x=1,y=0,vx=0,vy=2*np.pi)
os = sim.calculate_orbits()
print([os[0].a, os[0].e, os[0].omega, os[0].M])
rebound.OrbitPlot(sim,xlim=[-1,1],ylim=[-1,1])
```

#### Now get them by setting sim.G

```python
sim = rebound.Simulation()
sim.G  = 4*np.pi**2
sim.add(m=1,x=0,y=0,vx=0,vy=0)
sim.add(m=0,x=1,y=0,vx=0,vy=2*np.pi)
os = sim.calculate_orbits()
print([os[0].a, os[0].e, os[0].omega, os[0].M])
rebound.OrbitPlot(sim,xlim=[-1,1],ylim=[-1,1])
```
